```python
#1
m1 = list(map(int, input().split()))
m2 = list(map(int, input().split()))
#print(m1)
m = []
n1 = n2 = 0
while n1 < len(m1) and n2 < len(m2):
    if m1[n1] <= m2[n2]:
        m.append(m1[n1])
        n1 += 1
    else:
        m.append(m2[n2])
        n2 += 1

m += m1[n1:]
m += m2[n2:]
print(m)
```

```python
#2
m = []
n = 0
ln = 0
while(True):
    m1 = list(map(int, input().split()))
    if len(m1) == 1:
        n = m1[0]
        break
    else:
        for i in m1:
            m.append(i)
        ln += len(m1)
m.sort()
print(m[ln - n])
```

```python
#4
n = int(input())
while n:
    s1 = input()
    s2 = input()
    t1 = t2 = ""
    for i in s1:
        t1 += i.lower()
    for i in s2:
        t2 += i.lower()
    tmp = t = 0
    b = 0
    while t < len(s2):
        if t1[tmp] == t2[t]:
            tmp += 1
        if (tmp == len(s1)):
            b = 1
            break
        t += 1
    if b:
        print("YES")
    else:
        print("NO")
    n -= 1
```

```python
#6 next permu

n = int(input())
m = list(map(int, input().split()))

t = len(m) - 2
while t >= 0 and m[t] >= m[t+1]:
    t -= 1
if t >= 0:
    j = len(m) - 1
    while m[j] <= m[t]:
        j -= 1
    tmp = m[j]
    m[j] = m[t]
    m[t] = tmp
z = []
for i in range(t+1, len(m)):
    z.append(m[i])
z = z[::-1]
z = m[0:t+1] + z
print(z)
```

```python
#8
m = input()
dct = {}
dct[(0,0)] = 1
b = 0
x = y = 0
for i in m:
    if i == "U":
        y += 1
    elif i == "D":
        y -= 1
    elif i == "L":
        x -= 1
    elif i == "R":
        x += 1
    if (x,y) not in dct:
        dct[(x,y)] = 1
    else:
        b = 1
        break
if b:
    print("YES")
else:
    print("NO")
```
