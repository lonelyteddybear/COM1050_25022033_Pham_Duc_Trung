```python
#1
m1 = list(map(int, input().split()))
m2 = list(map(int, input().split()))
t1 = 0
t2 = 0
arr = []
while t1 < len(m1) and t2 < len(m2):
    if (m1[t1] <= m2[t2]):
        arr.append(m1[t1])
        t1 += 1
    else:
        arr.append(m2[t2])
        t2 += 1
arr += m1[t1:]
arr += m2[t2:]
print(arr)
```

```python
#7
def canPlaceFlowers(flowerbed, k):
    n = len(flowerbed)
    tot = 0
    mark = [0] * n
    for i in range(n):
        if (flowerbed[i] == 1):
            if (i == 0):
                mark[i+1] = 1
                mark[i] = 1
            elif (i == n-1):
                mark[i-1] = 1
                mark[i] = 1
            else:
                mark[i-1] = 1
                mark[i] = mark[i+1] = 1
    tmp = 0
    for i in range(n):
        if (mark[i] == 0):
            tmp += 1
        else:
            tot += (tmp + 1) // 2
    if (tmp > 0):
        tot += (tmp + 1) // 2
    if (tot >= k):
        return True
    return False

m = list(map(int, input().split()))
k = int(input())
if canPlaceFlowers(m,k):
    print(True)
else:
    print(False)
```

```python
#8
m = input()
dct = {}
dct[(0,0)] = 1
b = 0
x = y = 0
for i in m:
    if i == "U":
        y += 1
    elif i == "D":
        y -= 1
    elif i == "L":
        x -= 1
    elif i == "R":
        x += 1
    if (x,y) not in dct:
        dct[(x,y)] = 1
    else:
        b = 1
        break
if b:
    print("YES")
else:
    print("NO")
```

```python
#10
m = list(map(int, input().split()))

z = list(map(int, input().split()))

le = 0
ri = 1000000000
while le <= ri:
    mid = (le + ri) // 2

    ck = True
    for i in m:
        val = i
        b = 0
        for j in z:
            low = j - mid
            high = j + mid
            if low <= val <= high:
                b= 1
                break
        if not b:
            ck = False
            break
    if not ck:
        le = mid + 1
    else:
        ri = mid - 1

print(le)
```
