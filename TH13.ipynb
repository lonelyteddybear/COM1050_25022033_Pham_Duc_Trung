```python
from pathlib import Path
WORKDIR = Path(".")

path = "W11A2.txt"

def canWinNim (path: str) -> bool:
    file = WORKDIR/path
    with open (file, "r", "utf-8") as f:
        n = int(f.read().strip())

    if n % 4 == 0:
        return False
    return True

if (canWinNim(path)):
    print("WIN")
else:
    print("LOSE")
```

```python
from pathlib import Path
WORKDIR = Path(".")

path = "W11A5.txt"

def count (path: str):
    file = WORKDIR/path
    lst = []
    with open(file, "r") as f:
        for i in f:
            #n = str(i)
            lst.append(i.strip())
        #for i in range(1, len(lst)):
        #    print(lst[i])
        a = lst[1]
        s = lst[0]
        #print(s)
        cnt = 0
        for i in s:
            if i == a:
                cnt += 1
        return cnt
print(count(path))
```
