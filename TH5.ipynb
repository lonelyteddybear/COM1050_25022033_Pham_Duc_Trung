```python
def swap(a,b):
    m = a
    a = b
    b = m
    return (a,b)

m,n = map(int, input().split())
print(type(swap(m,n)))
```

```python
def prime(n):
    for i in range(2, int(n ** 0.5) + 1):
        if (n % i == 0):
            return False
    return n >= 2


n = int(input())
print(prime(n))
```

```python
import math as m
def chk_nto(a):
    if (a < 2):
        return False
    elif a == 2 or a == 3:
        return True
    for i in range(2, int(m.sqrt(a)) + 1):
        if (a % i == 0):
            return False
    return True
```

```python
def find(a, k):
    z = len(a)
    for i in range(z):
        if (a[i] == k):
            return i
    return -1
```

```python
def fact(n):
    if (n == 1):
        return 1
    return fact(n-1) * n
```

```python
def calc(num1, num2, x):
    if (x == "+"):
        tot = num1 + num2
        return (f'{tot:.2f}')
    elif (x == "-"):
        tot = num1 - num2
        return (f'{tot:.2f}')
    elif (x == "/"):
        if (int(num2) == 0):
            return ("IMPOSSIBLE")
        tot = num1 / num2
        return (f'{tot:.2f}')
    elif (x == "*"):
        tot = num1 * num2
        return (f'{tot:.2f}')
    else:
        return ("INVALID")
n1, n2 = map(float, input().split())
x = input()
print(calc(n1, n2, x))
```

```python
def sum_digit(n):
    tot = 0
    while n:
        tot += n % 10
        n // 10
    return tot
```

```python
def Hamming_Dis(n, m):
    n1 = ""
    m1 = ""
    while n > 0:
        tmp = n % 2
        t = chr(tmp + ord('0'))
        n1 = t + n1
        n //= 2
    while m > 0:
        tmp2 = m % 2
        t2 = chr(tmp2 + ord('0'))
        m1 = t2 + m1
        m //= 2
    #return (n1, m1)
    
    while (len(n1) > len(m1)):
        m1 = "0" + m1
    while (len(n1) < len(m1)):
        n1 = "0" + n1
    tot = 0
    for i in range(len(n1)):
        if (n1[i] != m1[i]):
            tot += 1
    return tot
    
n1, n2 = map(int, input().split())
#print(n1, n2)
print(Hamming_Dis(n1, n2))
```

```python
def dang_cau(m1, m2):
    chk = [-1] * 256
    l = len(m1)
    for i in range(l):
        tmp = ord(m1[i]) 
        tmp2 = ord(m2[i]) 
        if chk[tmp] == -1:
            chk[tmp] = tmp2
        else:
            if (tmp2 != chk[tmp]):
                return False

    chk = [-1] * 256
    l = len(m2)
    for i in range(l):
        tmp = ord(m2[i]) 
        tmp2 = ord(m1[i]) 
        if chk[tmp] == -1:
            chk[tmp] = tmp2
        else:
            if (tmp2 != chk[tmp]):
                return False
    return True

print(dang_cau("abca", "zbbz"))
#dang cau khi 1 ki tu thay = 1 ki tu khac, 2 ki tu khac nhau ko the cung toi 1 ki tu
```
