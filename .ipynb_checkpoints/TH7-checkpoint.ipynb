```python
#1
def binary_search(arr, left, right, target):
    while (left <= right):
        mid = (left + right) // 2
        if (arr[mid] == target):
            return(mid)
        elif arr[mid] > target:
            right = mid-1
        else:
            left = mid+1

    return(-1)
        
m = list(map(int, input().split()))
x = int(input())
print(binary_search(m, 0, len(m) - 1, x))
```

```python
#3
def bubble_sort(arr):
    n = len(arr)
    b = True
    while(b):
        change = 0
        for i in range(n-1):
            if (arr[i] > arr[i+1]):
                change = 1
                arr[i], arr[i+1] = arr[i+1], arr[i]
        if not change:
            b = False
    return arr

def selection_sort(arr):
    n = len(arr)
    for i in range(n-1):
        idx = i
        mn = arr[i]
        for j in range(i, n):
            if arr[j] < mn:
                idx = j
                mn = arr[j]
        arr[idx], arr[i] = arr[i], arr[idx]
    return arr

m = list(map(int, input().split()))

print(selection_sort(m))
print(bubble_sort(m))
```
