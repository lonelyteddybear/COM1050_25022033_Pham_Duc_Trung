```python
def prime(n):
    for i in range(2, int(n ** 0.5) + 1):
        if (n % i == 0):
            return False
    return n >= 2


n = int(input())
print(prime(n))
```

```python
#10
m = list(map(int, input().split()))
k = int(input())
dct = []
for i in range(len(m)):
    for j in range(i+1, len(m)):
        if(m[i] + m[j] == k):
            if (m[i] > m[j] and (m[j], m[i]) not in dct):
                dct.append((m[j], m[i]))
            elif  (m[i] < m[j] and (m[i], m[j]) not in dct):
                dct.append((m[i], m[j]))
dct.sort(key = lambda x: x[0]) #sorted by first element
for i in dct:
    print(i)
```

```python
#19
def max_pos(lst):
    mx = lst[0]
    arr = []
    for i in range(1, len(lst)):
        if mx < lst[i]:
            arr.clear()
            arr.append(i+1)
            mx = lst[i]
        elif mx == lst[i]:
            arr.append(i+1)
    return arr

m = list(map(int, input().split()))
res = max_pos(m)
if not len(res):
    print(-1)
else:
    print(*res)
```

```python
#22
def max_pos(lst):
    mx = lst[0]
    arr = []
    for i in range(1, len(lst)):
        if mx < lst[i]:
            arr.clear()
            arr.append(i+1)
            mx = lst[i]
        elif mx == lst[i]:
            arr.append(i+1)
    return arr

m = list(map(int, input().split()))
res = max_pos(m)
if not len(res):
    print(-1)
else:
    print(*res)
```

```python
#23
def erase_small(lst, pivot):
    cnt = 0
    while(cnt < len(lst)):
        if lst[cnt] < pivot:
            lst.pop(cnt)
        else:
            cnt += 1
    #print(cnt)
    return lst

m = list(map(int, input().split()))
x = int(input())
res = erase_small(m, x)
print(*res)
```

```python
#24
def push(lst):
    tmp1 = []
    tmp2 = []
    for i in lst:
        if not i:
            tmp1.append(0)
        else:
            tmp2.append(i)
    tmp1.extend(tmp2[0:])
    return tmp1

m = list(map(int, input().split()))
res = push(m)
print(*res)
```

```python
#27
def find_pos(lst, pivot):
    arr = []
    for i in range(0, len(lst)):
        if lst[i] < pivot:
            arr.append(i+1)
    return arr

m = list(map(int, input().split()))
k = int(input())
res = find_pos(m, k)
if not len(res):
    print(-1)
else:
    print(*res)
```

```python
#28

m = list(map(int, input().split()))
inc = 0
dec = 0
neutral = 0
for i in range(0, len(m) - 1):
    if (m[i] < m[i+1]):
        inc += 1
    elif m[i] > m[i+1]:
        dec += 1
    else:
        neutral = 1
        break
if (neutral or (inc and dec)):
    print("KHONG TANG KHONG GIAM")
elif dec == 0:
    print("TANG")
else:
    print("GIAM")
```
