```python
#1
def binary_search(arr, left, right, target):
    while (left <= right):
        mid = (left + right) // 2
        if (arr[mid] == target):
            return(mid)
        elif arr[mid] > target:
            right = mid-1
        else:
            left = mid+1

    return(-1)
        
m = list(map(int, input().split()))
x = int(input())
print(binary_search(m, 0, len(m) - 1, x))
```

```python
#3
def bubble_sort(arr):
    n = len(arr)
    b = True
    while(b):
        change = 0
        for i in range(n-1):
            if (arr[i] > arr[i+1]):
                change = 1
                arr[i], arr[i+1] = arr[i+1], arr[i]
        if not change:
            b = False
    return arr

def selection_sort(arr):
    n = len(arr)
    for i in range(n-1):
        idx = i
        mn = arr[i]
        for j in range(i, n):
            if arr[j] < mn:
                idx = j
                mn = arr[j]
        arr[idx], arr[i] = arr[i], arr[idx]
    return arr

m = list(map(int, input().split()))

print(selection_sort(m))
print(bubble_sort(m))
```

```python
#4
def find_occur(arr):
    dct = {}
    for i in arr:
        if i not in dct:
            dct[i] = 1
        else:
            dct[i] += 1
    mx = 0
    num = 0
    for i in dct:
        if (dct[i] > mx):
            mx = dct[i]
            num = i
    print(mx, num)
    return f'{num} xuat hien nhieu lan nhat, som nhat, {mx} lan'

m = list(map(int, input().split()))
print(find_occur(m))
```

```python
#5
m = list(map(int, input().split()))
x = int(input())

dct = {}
for i in m:
    if i not in dct:
        dct[i] = 1
    else:
        dct[i] += 1

sum = 0

for i in dct:
    if ((x - i) in dct):
        sum += dct[i] * dct[x - i]

print(sum // 2)
```

```python
#7
m = eval(input())
dp = [1] * (len(m) + 1)
for i in len(m):
    for j in range(0, i):
        if (m[i] > m[j]):
            dp[i] = max(dp[i], dp[j] + 1)
print(max(dp))
```
