```python
def count_occurrences(arr, x):
    cnt = 0
    for i in arr:
        if x == i:
            cnt += 1
    return cnt
```

```python
def bubble_sort(arr):
    n = len(arr)
    b = True
    while(b):
        change = 0
        for i in range(n-1):
            if (arr[i] > arr[i+1]):
                change = 1
                arr[i], arr[i+1] = arr[i+1], arr[i]
        if not change:
            b = False
    return arr

def selection_sort(arr):
    n = len(arr)
    for i in range(n-1):
        idx = i
        mn = arr[i]
        for j in range(i, n):
            if arr[j] < mn:
                idx = j
                mn = arr[j]
        arr[idx], arr[i] = arr[i], arr[idx]
    return arr

m = list(map(int, input().split()))

print(selection_sort(m))
print(bubble_sort(m))
```

```python
def find_occur(arr):
    dct = {}
    for i in arr:
        if i not in dct:
            dct[i] = 1
        else:
            dct[i] += 1
    mx = 0
    num = 0
    for i in dct:
        if (dct[i] > mx):
            mx = dct[i]
            num = i
    print(mx, num)

m = list(map(int, input().split()))
print(find_occur(m))
```

```python
m = list(map(int, input().split()))
x = int(input())

dct = {}
for i in m:
    if i not in dct:
        dct[i] = 1
    else:
        dct[i] += 1

sum = 0

for i in dct:
    if ((x - i) in dct):
        sum += dct[i] * dct[x - i]

print(sum // 2)
```
