```python
#8
#tach so (c1)
#reverse luon (c2)
n = int(input())
cnt = 0
flag = True
if (str(n) == str(n)[::-1]):
    flag = False
while(flag):
    n1 = str(n)
    n1 = n1[::-1]
    n += int(n1)
    print(n1, n)
    cnt += 1
    if (str(n) == str(n)[::-1]):
        break
print(f'So lan tao palindrome la {cnt}')
```

```python
#8
def rev(a):
    nw = 0
    while a:
        tmp = a % 10
        nw *= 10
        nw += tmp
        a //= 10
    return nw
n = int(input())
cnt = 0
flag = True
if (n == rev(n)):
    flag = False
while(flag):
    print(n, rev(n))
    n += rev(n)
    cnt += 1
    if (n == rev(n)):
        break
print(f'So lan tao palindrome la {cnt}')
```

```python
#9
#in cac so chinh phuong co cac chu so khac nhau va <= n
#do phuc tap: O(n * len^2)
import math as m
def check(n):
    n = str(n)
    n = sorted(n)
    for i in range(0, len(n)-1):
        if (n[i] == n[i+1]):
            return False
    return True
n = int(input())
for i in range(1, n+1):
    z = int(m.sqrt(i))
    if (z * z == i and check(i)):
        print(i, end = " ")
```

```python
#13
import math as m
try: 
    x,y = map(int, input("Nhap 2 so: ").split())
    sum_x = sum_y = 0
    for i in range(2, int(m.sqrt(x)+1)):
        if (x % i == 0):
            sum_x += i 
            sum_x += x//i
    for i in range(2, int(m.sqrt(y)+1)):
        if (y % i == 0):
            sum_y += i 
            sum_y += y//i
    sum_x += 1
    sum_y += 1
    print(sum_x, sum_y)
    if (int(sum_x) == int(y) and int(sum_y) == int(x)):
        print("true")
    else:
        print("false")
except ErrorValue:
    print("Nhap loi")
```

```python
#15
# so ga + cho = m, so chan ga + so chan cho = n
# 2x + 2y = 2m, 2x + 4y = n => 2y = n - 2m
m, n = map(int, input().split())
if (n < 2 * m):
    print("Invalid")
else:
    tmp = n - 2 * m
    if tmp % 2:
        print("Invalid")
    else:
        cho = tmp // 2
        ga = m - cho 
        print(f'So luong ga va cho lan luot la {ga} va {cho}')
```
